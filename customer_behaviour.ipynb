```python
df['Review Rating'] = df.groupby('Category')['Review Rating'].transform(lambda x : x.fillna(x.median()))
```

```python
df.isnull().sum()
```

```python
df['Age Group'] = df['Age'].apply(lambda x: 'young' if x <=20 else
 'young_adult' if x <=30 else
 'adult' if x <=50 else 
 'old')
```

```python
df.head(5)
```

```python
df.columns = df.columns.str.lower().str.replace(' ','_')
```

```python
print(df.columns.tolist())
```

```python
df.rename(columns={'purchase_amount_(usd)': 'purchase_amount'}, inplace=True)
```

```python
print(df.columns.tolist())
```

```python
df['payment_mode'] = df['payment_method'].apply(lambda x:'Digital' if x in ['Venmo','Credit Card','Debit Card','PayPal','Bank Transfer']  else
'cash')
```

```python
df.head(20)
```

```python
df['reviews'] = df['review_rating'].apply(lambda x: 'Excellent' if x > 4 else
'Good' if x > 3 else 'Average' if x > 2 else 'Poor')
```

```python
from sqlalchemy import create_engine
from sqlalchemy.engine import URL

# Step 1 — Connection
url = URL.create(
    drivername = "postgresql+psycopg2",
    username   = "postgres",
    password   = "qwerty@123",    # replace with your password
    host       = "localhost",
    port       = 5432,
    database   = "postgres"
)
engine = create_engine(url)

# Step 2 — Check your cleaned df before pushing
print("Shape:", df.shape)
print("Columns:", df.columns.tolist())
print(df.head(3))

# Step 3 — Push cleaned df to PostgreSQL
df.to_sql(
    name      = 'customer_shopping',
    con       = engine,
    if_exists = 'replace',
    index     = False
)

print("✅ Data pushed successfully!")
```

```python
# Run this in jupyter notebook
from sqlalchemy import inspect

inspector = inspect(engine)
tables = inspector.get_table_names()
print("Tables in PostgreSQL:", tables)
```
